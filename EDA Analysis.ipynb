```python
#Valence Ã¶zelliÄŸine gÃ¶re yeni kategorik sÃ¼tun oluÅŸturma

# Kategorilere gÃ¶re Valence deÄŸerlerini gruplara ayÄ±rarak yeni bir sÃ¼tun oluÅŸturma fonksiyonu
def categorize_valence(valence):
    if valence < 0.3:
        return 'Depresif'
    elif 0.3 <= valence < 0.5:
        return 'Slow'
    elif 0.5 <= valence < 0.7:
        return 'Pozitif'
    else:
        return 'Mutlu'

# 'Valence' Ã¶zelliÄŸini kategorilere ayÄ±ran yeni bir sÃ¼tun oluÅŸturma
youtube['Valence_Category'] = youtube['Valence'].apply(categorize_valence)
```

```python
#Youtube ÅŸarkÄ±larÄ±nÄ± Speechiness deÄŸerlerine gÃ¶re kategorize etme

# Kategorilere gÃ¶re Speechiness deÄŸerlerini gruplara ayÄ±ran yeni bir sÃ¼tun oluÅŸturma fonksiyonu
def categorize_speechiness(speechiness):
    if speechiness < 0.33:
        return 'MÃ¼zikal'
    elif 0.33 <= speechiness < 0.5:
        return 'Rap'
    elif 0.5 <= speechiness < 0.7:
        return 'KarÄ±ÅŸÄ±k'
    else:
        return 'KonuÅŸma'

# 'Speechiness' Ã¶zelliÄŸini kategorilere ayÄ±ran yeni bir sÃ¼tun oluÅŸturma
youtube['Speechiness_Category'] = youtube['Speechiness'].apply(categorize_speechiness)
```
