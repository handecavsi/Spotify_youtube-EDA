*Kolon Tanımları:*

- **Artist**: Şarkının veya videonun sanatçısı.

- **Track**: Şarkının adı.

- **Channel**: Videonun yayınlandığı YouTube kanalı.
- **Views**: Video görüntülenme sayısı.
- **Likes**: Video beğeni sayısı.
- **Comments**: Video yorum sayısı.
- **Description**: Video açıklaması.
- **Licensed**: Videonun lisanslı içeriği temsil edip etmediğini belirtir.

```python
#Youtube şarkılarının valence dağılımı

import matplotlib.pyplot as plt
import seaborn as sns

# Violin plot oluştur
plt.figure(figsize=(10, 6))
sns.violinplot(data=youtube, y='Valence', color='skyblue')

plt.title('Distribution of Valence')
plt.ylabel('Valence')

plt.grid(True)
plt.tight_layout()

# Grafiği göster
plt.show()


```

```python
#Valence özelliğine göre yeni kategorik sütun oluşturma

# Kategorilere göre Valence değerlerini gruplara ayırarak yeni bir sütun oluşturma fonksiyonu
def categorize_valence(valence):
    if valence < 0.3:
        return 'Depresif'
    elif 0.3 <= valence < 0.5:
        return 'Slow'
    elif 0.5 <= valence < 0.7:
        return 'Pozitif'
    else:
        return 'Mutlu'

# 'Valence' özelliğini kategorilere ayıran yeni bir sütun oluşturma
youtube['Valence_Category'] = youtube['Valence'].apply(categorize_valence)
```

```python
#Youtube şarkılarını Speechiness değerlerine göre kategorize etme

# Kategorilere göre Speechiness değerlerini gruplara ayıran yeni bir sütun oluşturma fonksiyonu
def categorize_speechiness(speechiness):
    if speechiness < 0.33:
        return 'Müzikal'
    elif 0.33 <= speechiness < 0.5:
        return 'Rap'
    elif 0.5 <= speechiness < 0.7:
        return 'Karışık'
    else:
        return 'Konuşma'

# 'Speechiness' özelliğini kategorilere ayıran yeni bir sütun oluşturma
youtube['Speechiness_Category'] = youtube['Speechiness'].apply(categorize_speechiness)
```
